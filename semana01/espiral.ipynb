```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Make a spiral with many turns
for N in [10, 100, 1000, 10000, 100000]:
    VOLTAS = N/10

    theta = np.linspace(0, VOLTAS*np.pi, N//2)
    x1 = VOLTAS*np.cos(theta)*theta
    y1 = VOLTAS*np.sin(theta)*theta
    x2 = VOLTAS*np.cos(theta)*(theta + np.pi)
    y2 = VOLTAS*np.sin(theta)*(theta + np.pi)

    x = np.concatenate([x1,x2[::-1]])
    y = np.concatenate([y1,y2[::-1]])
    np.savetxt(f"data/spiral_{x.shape[0]}.txt", np.c_[x, y], fmt="%.3f")
```

```python


plt.plot(x1, y1)
plt.plot(x2, y2)

# set aspect ratio to 1
plt.gca().set_aspect(1)

plt.show()

```

```python

for N in [10, 100, 1000, 10000, 100000, 1000000]:

    theta = np.linspace(0, 2*np.pi, N, endpoint=False)

    x = np.cos(theta)
    y = np.sin(theta)

    np.savetxt(f"data/regular_{x.shape[0]}.txt", np.c_[x, y], fmt="%.9e")
```
