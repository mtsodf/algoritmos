```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Make a spiral with many turns
theta = np.linspace(0, 10*np.pi, 60)
x1 = np.cos(theta)*theta
y1 = np.sin(theta)*theta

plt.plot(x1, y1)

x2 = np.cos(theta)*(theta + np.pi)
y2 = np.sin(theta)*(theta + np.pi)
plt.plot(x2, y2)

# set aspect ratio to 1
plt.gca().set_aspect(1)

plt.show()

```

```python
x = np.concatenate([x1,x2[::-1]])
y = np.concatenate([y1,y2[::-1]])
plt.plot(x,y)
plt.show()

np.savetxt(f"data/spiral.txt", np.c_[x, y], fmt="%.3f")
```
