```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("execucoes.txt", sep=";")
df["tipo"] = df["file"].apply(lambda x: x.split("/")[-1].split("_")[0])
df["tipo"] = df.tipo.apply(lambda x: "caixeiro" if x == "polygon" else x)
df
```

```python
fig, ax= plt.subplots()
sns.lineplot(data=df, x="size", y="time(s)", hue="tipo", ax=ax)
ax.plot([10, 100], [1e-3, 1e-1], color="black", ls="--", label="Linear", marker="o")
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("Tamanho PolÃ­gono")
ax.set_ylabel("Tempo de ExecuÃ§Ã£o (s)")
plt.savefig("../tex/figures/performance.png")
plt.show()
```
